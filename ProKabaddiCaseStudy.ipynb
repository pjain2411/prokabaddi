```python
#Player data for season 7

from selenium import webdriver

import pandas as pd

import numpy as np

from selenium.webdriver.chrome.options import Options

from bs4 import BeautifulSoup

import time

 

driver = webdriver.Chrome('chromedriver.exe')

driver.get('https://www.prokabaddi.com/stats/26-102-total-points-statistics')

time.sleep(5)

def fullpage():

    while True:

        try:

            loadMoreButton = driver.find_element_by_xpath('//*[@id="load_more"]/span')

            loadMoreButton.click()

            time.sleep(2)

        except Exception as e:

            print(e)

            break 

def extractdata(pagelist):

    fullpage()
    try:

        header_1 = driver.find_element_by_xpath('//*[@id="68c87f09-ea84-410b-bea4-8e8a18d5e9c7"]/component/div/div/div/div/section/div/div[1]/div/div[2]/div[1]/div/div[1]/div[1]/span').text

        header_2 = driver.find_element_by_xpath('//*[@id="68c87f09-ea84-410b-bea4-8e8a18d5e9c7"]/component/div/div/div/div/section/div/div[1]/div/div[2]/div[1]/div/div[2]/div[1]/span').text

        header_3 = driver.find_element_by_xpath('//*[@id="68c87f09-ea84-410b-bea4-8e8a18d5e9c7"]/component/div/div/div/div/section/div/div[1]/div/div[2]/div[1]/div/div[3]/div[1]/span').text

        header_4 = driver.find_element_by_xpath('//*[@id="68c87f09-ea84-410b-bea4-8e8a18d5e9c7"]/component/div/div/div/div/section/div/div[1]/div/div[2]/div[1]/div/div[4]/div[1]/span').text

        row=np.array([header_1,header_2,header_3,header_4])

        pagelist.append(row)
        print("headers*****************",pagelist)

        for i in range(2,170):
            try:

                rank = driver.find_element_by_xpath('//*[@id="68c87f09-ea84-410b-bea4-8e8a18d5e9c7"]/component/div/div/div/div/section/div/div[1]/div/div[2]/div[%s]/div/div[1]/span' % i).text

                name = driver.find_element_by_xpath('//*[@id="68c87f09-ea84-410b-bea4-8e8a18d5e9c7"]/component/div/div/div/div/section/div/div[1]/div/div[2]/div[%s]/div/a[1]/div[1]/span' % i).text

                player_type = driver.find_element_by_xpath('//*[@id="68c87f09-ea84-410b-bea4-8e8a18d5e9c7"]/component/div/div/div/div/section/div/div[1]/div/div[2]/div[%s]/div/a[1]/div[2]/span' % i).text       

                team = driver.find_element_by_xpath('//*[@id="68c87f09-ea84-410b-bea4-8e8a18d5e9c7"]/component/div/div/div/div/section/div/div[1]/div/div[2]/div[%s]/div/a[2]' % i).get_attribute('href')

                matches_played = driver.find_element_by_xpath('//*[@id="68c87f09-ea84-410b-bea4-8e8a18d5e9c7"]/component/div/div/div/div/section/div/div[1]/div/div[2]/div[%s]/div/div[2]/span' % i).text

                total_points = driver.find_element_by_xpath('//*[@id="68c87f09-ea84-410b-bea4-8e8a18d5e9c7"]/component/div/div/div/div/section/div/div[1]/div/div[2]/div[%s]/div/div[3]/span' % i).text

                row=np.array([rank,name,player_type,team,matches_played,total_points])

                pagelist.append(row)
            except Exception as e:
 
                print(e)
                break

        print("*****************",pagelist)

        return pagelist

    except Exception as e:

             print(e)

#select drop down

driver.find_element_by_xpath('//*[@id="season_dropdown"]/option[2]').click()

time.sleep(2)

for n in range(1,13):

    driver.find_element_by_xpath('//*[@id="si_dropdown"]/option[%s]' % n).click()

    time.sleep(2)

    pagelist=[]

    pagelist = extractdata(pagelist)

    df=pd.DataFrame(pagelist)

    export_csv = df.to_csv (r'season7_players_%s.csv' % n, header=True)     

    
          

```
